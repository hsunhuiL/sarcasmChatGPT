```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
df = pd.read_csv("/kaggle/input/data/Dataset.csv")
P_measures = df[["Familiarity_P","Imageability_P","Meaningfulness_P","Brysbaert_P","comp_p","WC_p","read_p"]]
R_measures = df[["Familiarity_R","Imageability_R","Meaningfulness_R","Brysbaert_R","comp_r","WC_r","read_r"]]
```

```python
# plot heatmap for correlation in Prompt's feature

plt.figure(figsize=(9,6))
sns.heatmap(P_measures.corr(method='spearman'), cmap='coolwarm', annot=True)
plt.tight_layout()
plt.subplots_adjust(top=0.85)
plt.savefig("prompt_corr.png")
plt.show()
```

```python
# plot heatmap for correlation in Response's feature

plt.figure(figsize=(9,6))
sns.heatmap(R_measures.corr(method='spearman'), cmap='coolwarm', annot=True)
plt.savefig("res_corr.png")
plt.show()
```

Imageability is highly correlated with Meaningfulness and Concreteness, in both prompts and responses
